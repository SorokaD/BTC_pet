### 1. **Сбор данных: получение цен через WebSocket API Bybit**

Bybit WebSocket API позволяет получать обновления по ценам в реальном времени. Для интеграции можно использовать Python-библиотеку `websockets` или библиотеку `bybit`, если она поддерживает нужные функции.

```python
import asyncio
import websockets
import json
import redis

# Инициализация Redis для хранения данных
redis_client = redis.Redis(host='localhost', port=6379, db=0)

async def subscribe_to_price_stream():
    url = "wss://stream.bybit.com/realtime"  # URL WebSocket API Bybit
    async with websockets.connect(url) as ws:
        # Подписка на канал котировок, например BTC/USDT
        await ws.send(json.dumps({
            "op": "subscribe",
            "args": ["trade.BTCUSDT"]
        }))
        
        while True:
            response = await ws.recv()
            data = json.loads(response)
            
            # Обработка данных и сохранение в Redis
            if 'data' in data:
                price = data['data'][0]['price']
                redis_client.set("BTCUSDT_price", price)  # Сохранение в Redis
                print(f"Price: {price}")

asyncio.run(subscribe_to_price_stream())
```

Этот скрипт будет каждую секунду получать обновление по цене BTCUSDT и сохранять его в Redis.

### 2. **Обработка данных и запуск моделей**

После получения данных вам нужно создать модуль, который будет загружать предобученные модели и запускать их на основе полученных данных.

#### Заготовка для ML модели

Предположим, что у вас есть предобученная модель, сохранённая как файл (например, `model.pkl`). Вы можете использовать библиотеку `joblib` для загрузки модели.

```python
import joblib
import numpy as np

# Загрузка предобученной модели
model = joblib.load("model.pkl")

def run_model(price):
    # Обработка цены, приведение к необходимому формату
    features = np.array([price]).reshape(1, -1)
    
    # Предсказание модели
    decision = model.predict(features)
    return decision  # Например, 1 - покупка, -1 - продажа, 0 - удержание
```

### 3. **Торговый модуль: выполнение торговых решений через REST API Bybit**

Этот модуль использует REST API Bybit для создания или закрытия ордеров. Подключение к API можно сделать с помощью стандартных HTTP-запросов, или библиотеки `bybit`.

```python
import requests
import json
import time

API_KEY = "your_api_key"
SECRET_KEY = "your_secret_key"
BASE_URL = "https://api.bybit.com"  # Базовый URL Bybit API

def place_order(symbol, side, qty, price):
    endpoint = "/v2/private/order/create"
    headers = {
        'Content-Type': 'application/json',
        'X-BYBIT-APIKEY': API_KEY
    }
    data = {
        "symbol": symbol,
        "side": side,
        "order_type": "Limit",
        "qty": qty,
        "price": price,
        "time_in_force": "GoodTillCancel"
    }
    response = requests.post(BASE_URL + endpoint, headers=headers, data=json.dumps(data))
    return response.json()

# Пример вызова функции для создания ордера на покупку
response = place_order("BTCUSDT", "Buy", 0.01, 30000)
print(response)
```

### 4. **Мониторинг и логирование**

Для мониторинга состояния системы и логирования результатов можно использовать Prometheus и Grafana.

#### Логирование в Prometheus с использованием `prometheus_client`

```python
from prometheus_client import start_http_server, Gauge
import time

# Инициализация метрик
price_metric = Gauge('crypto_price', 'Price of the cryptocurrency', ['currency'])

# Запуск HTTP сервера Prometheus
start_http_server(8000)

while True:
    # Получаем цену из Redis
    price = redis_client.get("BTCUSDT_price")
    if price:
        price_metric.labels(currency="BTCUSDT").set(float(price))
    time.sleep(1)
```

Теперь вы можете настроить Grafana для отображения метрики `crypto_price`.
